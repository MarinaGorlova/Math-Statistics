# Изучение функций с помощью MatplotLib

В Python 3

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Цель - исследование функции 
$$\LARGE f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$$
при различных значениях $\alpha$ и $\beta$.

Для каждой задачи необходимо:

- Построить график (размер графика должен быть достаточным, чтобы визуально увидеть особенности изучаемых функций), график каждой функции должен быть одного цвета для одного значения $\alpha$ и $\beta$.

- Подписать оси и заголовок

- Создать легенду

- Сохранить изображение в svg файл

- Сделать и записать выводы о поведении графиков


## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
def f(x): 
    return (x + 1)/(x)
def g(x):
    return (x + 2)/(x)
def h(x):
    return (x**2 + 1)/(x**2)
def y(x):
    return 0*x

x = np.ma.masked_inside(np.linspace(-6., 6., num=500), -0.02, 0.02)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)
axes.plot(x, f(x), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes.plot(x, g(x), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes.plot(x, h(x), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes.set_xlabel('x', fontsize=14)
axes.set_ylabel('f(x)', fontsize=14)
axes.set_title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$', fontsize=16)
axes.set_ylim(ymin = -30, ymax = 30)
fig.tight_layout()                     
axes.legend() 
plt.savefig('1.1.svg')

plt.show()
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.linspace(0.01, 6, num=500)
x1 = np.linspace(0.4, 1.4, num=1000)
x2 = np.linspace(1000, 1120, num=500)

fig, axes = plt.subplots(figsize=(13,8), dpi=120) 

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f(x), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes1.plot(x, g(x), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes1.plot(x, h(x), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes1.set_xlabel('x', fontsize=14)
axes1.set_ylabel('f(x)', fontsize=14)
axes1.set_title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$ для x>0', fontsize=16)
axes1.set_ylim(ymin = 0, ymax = 30)                 
axes1.legend() 

axes2 = fig.add_axes([0.125, 0.5, 0.35, 0.35])
axes2.plot(x1, f(x1), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes2.plot(x1, g(x1), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes2.plot(x1, h(x1), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes2.set_xlabel('x', fontsize=10)
axes2.set_ylabel('f(x)', fontsize=10)
axes2.set_title('для малых x')
axes2.set_ylim(ymin = 1, ymax = 7)    
axes2.legend() 

axes3 = fig.add_axes([0.55, 0.5, 0.35, 0.35])
axes3.plot(x2, f(x2), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes3.plot(x2, g(x2), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes3.plot(x2, h(x2), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes3.set_xlabel('x', fontsize=10)
axes3.set_ylabel('f(x)', fontsize=10)
axes3.set_title('для больших x')
axes3.set_ylim(ymin = 0.998, ymax = 1.004)
axes3.legend()           

plt.savefig('1.2.svg')
plt.show()
```

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.linspace(-6, -0.01, num=500)
x1 = np.linspace(-1110, -1000, num=500)

fig, axes = plt.subplots(figsize=(13,8), dpi=120) 

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f(x), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes1.plot(x, g(x), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes1.plot(x, h(x), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes1.plot(x, y(x), 'k', label = 'y(x) = 0', lw=2.5)
axes1.set_xlabel('x', fontsize=14)
axes1.set_ylabel('f(x)', fontsize=14)
axes1.set_title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$ для x<0', fontsize=16)
axes1.set_ylim(ymin = -30, ymax = 30)        
axes1.legend() 

axes2 = fig.add_axes([0.3, 0.6, 0.35, 0.35])
axes2.plot(x1, f(x1), 'r', label = '$\\alpha = 1, \\beta = 1$')  
axes2.plot(x1, g(x1), 'g', label = '$\\alpha = 2, \\beta = 1$')
axes2.plot(x1, h(x1), 'b', label = '$\\alpha = 1, \\beta = 2$')
axes2.set_xlabel('x', fontsize=10)
axes2.set_ylabel('f(x)', fontsize=10)
axes2.set_title('при удалении x от 0 к −$\\infty$')
axes2.set_ylim(ymin = 0.996, ymax = 1.002)   
axes2.legend()

plt.savefig('1.3.svg')
plt.show()
```

Сделайте выводы о поведении графиков при различных значениях $\alpha$ и $\beta$.

Изучите возможности пересечения и то к чему стремятся кривые.

Сделайте выводы о возрастании/убывании.

Значения 1-й функции (при $\alpha$ = 1 и $\beta$ = 1, красная линия) стремятся к +$\infty$ и к -$\infty$ при х, стремящемся к 0. Функция принимает как положительные, так и отрицательные значения.<br> 
Значения 2-й функции (при $\alpha$ = 2 и $\beta$ = 1, зелёная линия) стремятся к +$\infty$ и к -$\infty$ при х, стремящемся к 0. Функция принимает как положительные, так и отрицательные значения.<br> 
Значения 3-й функция (при $\alpha$ = 1 и $\beta$ = 2, синяя линия) стремятся к +$\infty$ при х, стремящемся к 0. Функция принимает только положительные значения.<br>
График ни одной из трёх фнукций не пересекает ось y. Все 3 функции терпят разрыв при х = 0. х = 0 является асимптотой для всех трёх функций, при стремлении к которой слева первые 2 функции неограниченно монотонно убывают, а 3-я - возрастает.<br> 
При стремлении к x=0 справа все 3 функции неограниченно монотонно возрастают.<br>
На графике функций при x>0 для малых x заметны 2 точки пересечения: при х = 0.5 (2-я и 3-я функции принимают значение = 5) и при х = 1 (1-я и 3-я функции принимают значение = 2).<br>
При стремлении x к +$\infty$ все 3 функции стремятся к значению, равному 1.<br>    
На графике функций при x<0 при стремлении x к −$\infty$ заметно, что все 3 функции стремятся к значению, равному 1.<br>

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
def q(x, alpha, beta): 
    return (x**beta + alpha**beta)/(x**beta)

x = np.linspace(-6., 6., num=500)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)
axes.plot(x, q(x, 1, 0), 'r', label = '$\\alpha = 1, \\beta = 0$')  
axes.plot(x, q(x, 0, 0), 'g', label = '$\\alpha = 0, \\beta = 0$')
axes.plot(x, q(x, 0, 1), 'b', label = '$\\alpha = 0, \\beta = 1$')
axes.plot(x, q(x, 0, 2), 'c', label = '$\\alpha = 0, \\beta = 2$')  
axes.plot(x, q(x, 1, -1), 'm', label = '$\\alpha = 1, \\beta = -1$')
axes.plot(x, q(x, 2, -1), 'y', label = '$\\alpha = 2, \\beta = -1$')
axes.set_xlabel('x', fontsize=14)
axes.set_ylabel('f(x)', fontsize=14)
axes.set_title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$', fontsize=16)
axes.set_ylim(ymin = -5, ymax = 5)
fig.tight_layout()                     
axes.legend() 
plt.savefig('2.svg')
plt.show()
```

Графики первых двух функций совпадают и принимают значения, равные 2, при любых значениях х.<br> 
Графики третьей и четвертой функций совпадают и принимают значения, равные 1, при любых значениях х.<br> 
График пятой функции (розовая линия) является линейной функцией f(x) = x + 1. Эта функция смещена вверх по оси ординат на 1. При увеличении х она неограниченно возрастает, при уменьшении x - неограниченно убывает.<br>
График шестой функции (оливковая линия) является линейной функцией f(x) = 0,5x + 1. Ведет себя так же, как 5-я функция. Отличается от нее меньшим углом наклона к оси абсцисс. <br>
Все функции имеют в целом 3 точки пересечения: при x = 0 (четыре последние функции), при х = 1 (5-я функция и две первых, при х = 2 (6-я функции и 2 первых).

## Задача №3

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x = np.ma.masked_inside(np.linspace(-6., 6., num=5000), -0.02, 0.000002)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)
axes.plot(x, q(x, 1, 0.5), 'r', label = '$\\alpha = 1, \\beta = 0.5$')  
axes.plot(x, q(x, 1, -0.5), 'g', label = '$\\alpha = 1, \\beta = -0.5$')
axes.plot(x, q(x, 1, -1.5), 'b', label = '$\\alpha = 1, \\beta = -1.5$')
axes.set_xlabel('x', fontsize=14)
axes.set_ylabel('f(x)', fontsize=14)
axes.set_title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$', fontsize=16)
axes.set_ylim(ymin = 0, ymax = 10)
fig.tight_layout()                     
axes.legend() 
plt.savefig('3.svg')
plt.show()
```

Первая функция (красная линия) не определена при неположительных значениях.<br>
Значение первой функции при стремлении х к 0 справа стремится к +$\infty$, т.е. неограниченно возрастает. Т.о., х = 0 является вертикальной асимпотой этой функции.<br>
При стремлении х к +$\infty$, значение функции стремятся к 1 (горизонтальная асимптота).<br>
Первая функция является выпуклой.<br>
<br>
Вторая функции (зеленая линия) не определена при отрицательных значениях.<br>
При стремлении х к +$\infty$, функция монотонно неограниченно возрастает.<br>
Вторая функция является вогнутой.<br>
<br>
Третья функция (зеленая линия) не определена при отрицательных значениях.<br>
При стремлении х к +$\infty$, функция монотонно неограниченно возрастает.<br>
Третья функция является выпуклой. <br>
<br>
Вторая и третья функция пересекаются в точке (0,1).<br>
Все 3 функции пересекаются в точке (1,2).<br>
